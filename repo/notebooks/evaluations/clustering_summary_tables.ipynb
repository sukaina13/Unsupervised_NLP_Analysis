#### Overall Clustering Performance Aggregation and Ranking

```python
# LOAD + AGGREGATE PER DATASET
# Map dataset names to their result files
datasets_main = {
    "Amazon": "../../results/clustering/benchmark/amazon_clustering_scores.csv",
    "arXiv": "../../results/clustering/benchmark/arxiv_clustering_scores.csv",
    "RCV1": "../../results/clustering/benchmark/rcv1_clustering_scores.csv",
    "WOS": "../../results/clustering/benchmark/wos_clustering_scores.csv",
    "DBpedia": "../../results/clustering/benchmark/db_clustering_scores.csv"}

# Store summarized results for each dataset
results_main = {}
for name, file in datasets_main.items():
    # Load clustering results
    df = pd.read_csv(file)

    # Remove "level" column if present
    df = df.drop(columns=["level"], errors="ignore")

    # Key step: aggregate multiple runs into one stable result
    # Group by embedding, reduction, and clustering method
    # Then take the median to reduce noise
    df_summary = df.groupby(
        ["embedding_model", "reduction_method", "cluster_method"]
    ).median(numeric_only=True)
    # Save results for this dataset
    results_main[name] = df_summary

# COMBINE ALL DATASETS
# Merge all dataset results into one table
all_results = pd.concat([df.reset_index() for df in results_main.values()],
    ignore_index=True)


#FILTER METHODS

# Keep only valid dimensionality reduction methods
valid_reductions = ["PCA", "UMAP", "PHATE", "PaCMAP", "tSNE"]

all_results = all_results[
    all_results["reduction_method"].isin(valid_reductions)
]


# TWO-STAGE AGGREGATION (IMPORTANT)
# First: aggregate per embedding model
# This ensures one embedding doesn't dominate the results
per_embedding = all_results.groupby(
    ["embedding_model", "reduction_method", "cluster_method"]
).median(numeric_only=True).reset_index()

# Second: aggregate across embeddings and datasets
# Produces one final score per configuration
overall = per_embedding.groupby(
    ["reduction_method", "cluster_method"]
).median(numeric_only=True).reset_index()


# CREATE READABLE LABELS
# Combine reduction + clustering into one name (e.g., "UMAP + Agglomerative")
overall["config"] = (
    overall["reduction_method"] + " + " + overall["cluster_method"])


# RANK METHODS
# Rank configurations based on clustering performance metrics
overall["ARI_rank"] = overall["ARI"].rank(ascending=False, method="average")
overall["AMI_rank"] = overall["AMI"].rank(ascending=False, method="average")
overall["FM_rank"]  = overall["FM"].rank(ascending=False, method="average")


# SORT RESULTS
# Sort by best overall performance (ARI first, then AMI)
overall = overall.sort_values(by=["ARI", "AMI"], ascending=False)
#FINAL TABLE
# Select key columns for reporting
final_table = overall[[
    "config", "ARI", "AMI", "FM", "ARI_rank", "AMI_rank", "FM_rank"]]

# Display final results
print("\n===== FINAL RESULTS TABLE =====")
print(final_table)
```

#### Dimensionality Reduction Evaluation on Synthetic Datasets
Synthetic datasets are generated with varying hierarchical depth (d3 vs d5) and noise levels (clean, 25%, 50%) to evaluate robustness of dimensionality reduction methods.

```python
import pandas as pd
import os

BASE = "../../results/viz_metrics"
MINILM_DIR = f"{BASE}/sentence-transformers"
QWEN_DIR   = f"{BASE}/Qwen"

CONFIGS = [
    ("Energy d3",    "Energy_Ecosystems_and_Humans",             "maxsub5_depth3"),
    ("Energy d5",    "Energy_Ecosystems_and_Humans",             "maxsub3_depth5"),
    ("Fisheries d3", "Offshore_energy_impacts_on_fisheries",     "maxsub5_depth3"),
    ("Fisheries d5", "Offshore_energy_impacts_on_fisheries",     "maxsub3_depth5"),
]

NOISE_MAP = {"Clean": "", "25%": "_noise0.25", "50%": "_noise0.5"}

def load_synthetic(base_dir, theme, config, noise_suffix):
    fname = f"viz_metrics_{theme}_hierarchy_t1.0_{config}_synonyms0{noise_suffix}_random.csv"
    df = pd.read_csv(os.path.join(base_dir, fname))
    return df.rename(columns={"Spearman Correlation": "Spearman"})[
        ["Method", "Trustworthiness", "Continuity", "Spearman", "DEMaP"]
    ]

rows = []
for label, theme, config in CONFIGS:
    for noise_label, noise_suffix in NOISE_MAP.items():
        minilm = load_synthetic(MINILM_DIR, theme, config, noise_suffix)
        qwen   = load_synthetic(QWEN_DIR,   theme, config, noise_suffix)
        merged = minilm.merge(qwen, on="Method", suffixes=("_MiniLM", "_Qwen"))
        merged.insert(0, "Condition", noise_label)
        merged.insert(0, "Dataset",   label)
        rows.append(merged)

df = pd.concat(rows, ignore_index=True).rename(columns={
    "Trustworthiness_MiniLM": "T_MiniLM", "Continuity_MiniLM": "C_MiniLM",
    "Spearman_MiniLM": "Sp_MiniLM",
    "Trustworthiness_Qwen": "T_Qwen", "Continuity_Qwen": "C_Qwen",
    "Spearman_Qwen": "Sp_Qwen",
})

summary_minilm = (
    df.groupby("Method")[["T_MiniLM", "C_MiniLM", "Sp_MiniLM", "DEMaP_MiniLM"]]
      .agg(["mean", "std"]).round(4))

summary_qwen = (
    df.groupby("Method")[["T_Qwen", "C_Qwen", "Sp_Qwen", "DEMaP_Qwen"]]
      .agg(["mean", "std"]).round(4))

def format_summary(summary_df, prefix):
    out = pd.DataFrame(index=summary_df.index)
    metric_names = {
        f"T_{prefix}": "Trustworthiness", f"C_{prefix}": "Continuity",
        f"Sp_{prefix}": "Spearman",       f"DEMaP_{prefix}": "DEMaP",
    }
    for old, new in metric_names.items():
        out[new] = (summary_df[(old, "mean")].map(lambda x: f"{x:.4f}")
                    + " +/- "
                    + summary_df[(old, "std")].map(lambda x: f"{x:.4f}"))
    return out.reset_index()

formatted_minilm = format_summary(summary_minilm, "MiniLM")
formatted_qwen   = format_summary(summary_qwen,   "Qwen")

print("Formatted MiniLM:")
print(formatted_minilm.to_string(index=False))
print("\nFormatted Qwen:")
print(formatted_qwen.to_string(index=False))

formatted_minilm.to_csv("../../results/clustering/synthetic/synthetic_dr_summary_minilm.csv", index=False)
formatted_qwen.to_csv("../../results/clustering/synthetic/synthetic_dr_summary_qwen.csv", index=False)
```

#### Dimensionality Reduction Performance on Benchmark Datasets

```python
import pandas as pd
import os

BASE = "../../results/viz_metrics/sentence-transformers"

DATASETS = {
    "RCV1":    "rcv1",
    "arXiv":   "arxiv",
    "Amazon":  "amazon",
    "DBpedia": "dbpedia",
    "WoS":     "wos",
}

dfs = []
for label, name in DATASETS.items():
    d = pd.read_csv(f"{BASE}/viz_metrics_{name}.csv")
    d["Dataset"] = label
    dfs.append(d)

df = pd.concat(dfs, ignore_index=True).rename(columns={
    "Trustworthiness_mean":      "Trustworthiness",
    "Continuity_mean":           "Continuity",
    "Spearman_Correlation_mean": "Spearman",
    "DEMaP_mean":                "DEMaP",
})[["Dataset", "Method", "Trustworthiness", "Continuity", "Spearman", "DEMaP"]]

summary = (
    df.groupby("Method")[["Trustworthiness", "Continuity", "Spearman", "DEMaP"]]
      .agg(["mean", "std"]).round(4))

def format_summary(summary_df):
    out = pd.DataFrame(index=summary_df.index)
    for metric in ["Trustworthiness", "Continuity", "Spearman", "DEMaP"]:
        out[metric] = (summary_df[(metric, "mean")].map(lambda x: f"{x:.4f}")
                       + " +/- "
                       + summary_df[(metric, "std")].map(lambda x: f"{x:.4f}"))
    return out.reset_index()

formatted = format_summary(summary)
print(formatted.to_string(index=False))
```
